```python
#Label_Encoding
from sklearn.preprocessing import LabelEncoder
a=LabelEncoder()
columns=df.select_dtypes(include=['object'])
for col in columns:
  df[col]=a.fit_transform(df[col])
df
```

```python
x=df.drop('Feedback',axis=1)
y=df['Feedback']
```

```python
#training_and_testing
from sklearn.model_selection import  train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
x_train
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler
f=StandardScaler()
x_train=f.fit_transform(x_train)
x_test=f.fit_transform(x_test)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model=keras.Sequential([keras.layers.Flatten(input_shape=(12,)),
                       keras.layers.Dense(256,activation='relu'),
                       keras.layers.Dense(128,activation='relu'),
                       keras.layers.Dense(2,activation='sigmoid')])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history=model.fit(x_train,y_train,validation_split=0.1,epochs=10)
```

```python
#accuracy_and_val_accuracy
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.xlabel('accuracy')
plt.ylabel('epochs')
plt.legend(['training_data','validation_data'],loc='lower right')
plt.show()
```

```python
#loss_and_val_loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.xlabel('loss')
plt.ylabel('epochs')
plt.legend(['training_data','validation_data'],loc='upper right')
plt.show()
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_pred=np.argmax(y_pred,axis=1)
y_pred
```

```python
from sklearn.metrics import accuracy_score
#accuracy_score
accuracy_score(y_pred,y_test)
```
